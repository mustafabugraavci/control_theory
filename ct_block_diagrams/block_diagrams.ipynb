```python
import control
import matplotlib.pyplot as plt

# Define system parameters
τ = 2.0
K = 3.0

# Create the system
numerator = [K]
denominator = [τ, 1]
system = control.TransferFunction(numerator, denominator)

# Calculate and plot the system's Bode plot
control.bode_plot(system)
plt.title('Bode Plot of a First-Order System')
plt.grid(True)
plt.show()
```
