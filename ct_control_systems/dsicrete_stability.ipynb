```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# Define the numerator and denominator of the transfer function
numerator = [1]
denominator = [1, -0.5, 0.75]

# Create a transfer function
system = signal.TransferFunction(numerator, denominator, dt=1)

# Calculate poles of the system
poles = np.roots(denominator)

# Determine stability
stable = all(abs(pole) < 1 for pole in poles)

# Print the results
print(f"Poles of the system: {poles}")
print(f"The system is {'stable' if stable else 'unstable'}.")

# Plot the poles on the z-plane
plt.figure()
plt.scatter(np.real(poles), np.imag(poles), marker='x')
plt.axhline(0, color='black', lw=1)
plt.axvline(0, color='black', lw=1)
plt.title("Pole-Zero Plot")
plt.xlabel("Real")
plt.ylabel("Imaginary")
plt.grid(True)
plt.show()
```
