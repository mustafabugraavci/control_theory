```python
import numpy as np
import matplotlib.pyplot as plt

# System parameters
T = 10  # Time constant of the system
Kp = 2  # Proportional gain
Ki = 1  # Integral gain
Kd = 0.5  # Derivative gain

# Initialize variables
setpoint = 50  # Desired temperature
time = np.arange(0, 100, 0.1)
error_integral = 0
prev_error = 0
output = []

# Closed-loop control simulation
for t in time:
    current_temp = 30 + np.random.randn()  # Simulated temperature with noise
    error = setpoint - current_temp
    error_integral += error
    error_derivative = error - prev_error
    control_signal = Kp * error + Ki * error_integral + Kd * error_derivative
    output.append(current_temp + control_signal)
    prev_error = error

# Plot results
plt.figure(figsize=(10, 6))
plt.plot(time, output, label='Temperature')
plt.xlabel('Time')
plt.ylabel('Temperature (°C)')
plt.title('Temperature Control System')
plt.legend()
plt.grid(True)
plt.show()
```
