```python
# Python code example: A simple linear ODE
from sympy import symbols, Eq, Function, Derivative

t = symbols('t')
y = Function('y')(t)
ode = Eq(Derivative(y, t), 3 * y - 2 * t)

# Solving the ODE
from sympy import dsolve

solution = dsolve(ode)
print(solution)
```

```python
# Python code example: Solving a simple nonlinear ODE numerically using SciPy
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Define the differential equation as a function
def odefunc(y, t):
    return y**2 - t**2

# Initial condition
y0 = 0.5

# Time points
t = np.linspace(0, 2, 100)

# Solve the ODE numerically
sol = odeint(odefunc, y0, t)

# Plot the solution
plt.plot(t, sol)
plt.xlabel('Time')
plt.ylabel('Solution y(t)')
plt.title('Numerical Solution of a Nonlinear ODE')
plt.grid(True)
plt.show()
```

```python
# Python code example: A first-order linear equation of state
from sympy import symbols, Eq, Function, Derivative

t = symbols('t')
x = Function('x')(t)
ode = Eq(Derivative(x, t), -2 * x + 5)

# Solving the equation
from sympy import dsolve

solution = dsolve(ode)
print(solution)
```
