```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# System parameters
m = 1.0  # Mass (kg)
c = 0.2  # Damping coefficient (N/m/s)
k = 1.0  # Spring constant (N/m)

# Input function (step input)
def input_function(t):
    return 1.0 if t >= 1.0 else 0.0

# System dynamics
def system_dynamics(t, z):
    x, x_dot = z
    u = input_function(t)
    x_dot_dot = (-c * x_dot - k * x + u) / m
    return [x_dot, x_dot_dot]

# Initial conditions
initial_state = [0.0, 0.0]

# Time span
t_span = (0, 10)

# Time points for integration
t_eval = np.linspace(t_span[0], t_span[1], 1000)

# Solve the ODE manually
sol = solve_ivp(system_dynamics, t_span, initial_state, t_eval=t_eval)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(sol.t, sol.y[0], label='Position (x)')
plt.plot(sol.t, sol.y[1], label='Velocity (x_dot)')
plt.xlabel('Time')
plt.ylabel('System Response')
plt.legend()
plt.grid(True)
plt.title('Mass-Spring-Damper System Response')
plt.show()
```

```python
import numpy as np
from scipy.linalg import solve_continuous_are

# Define the system matrix A (modify as needed)
A = np.array([[-2, 1], [3, -4]])

# Define the input matrix B (modify as needed)
B = np.array([[0], [1]])

# Define the cost matrices Q and R (modify as needed)
Q = np.eye(2)  # State cost matrix
R = np.eye(1)  # Input cost matrix

# Solve the continuous-time algebraic Riccati equation
P = solve_continuous_are(A, B, Q, R)

# Calculate the controller gain matrix K
K = np.linalg.inv(R) @ np.transpose(B) @ P

print("Controller Gain Matrix K:")
print(K)
```
