```python
import numpy as np
import matplotlib.pyplot as plt

# Define system parameters
τ = 2.0
K = 3.0

# Define time values
t = np.linspace(0, 10, 1000)

# Calculate impulse response
h = K * (1 / τ) * np.exp(-t / τ)

# Plot the impulse response
plt.figure(figsize=(8, 6))
plt.plot(t, h, label='Impulse Response')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Impulse Response of a First-Order System')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import control

# Define system parameters
m = 1.0
c = 0.5
k = 2.0

# Create the system
numerator = [1]
denominator = [m, c, k]
system = control.TransferFunction(numerator, denominator)

# Calculate and plot the system's Bode plot
control.bode_plot(system)
plt.title('Bode Plot of a Second-Order Mechanical System')
plt.grid(True)
plt.show()
```
