```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the dead-time system
dead_time = 2  # Time delay in seconds
sampling_time = 0.1  # Sampling time in seconds

# Define the dead-time system transfer function
def dead_time_system(s):
    return np.exp(-dead_time * s)

# Generate frequency response
frequency = np.linspace(0, 2, 1000)  # Frequency range from 0 to 2 Hz
s = 1j * 2 * np.pi * frequency  # Laplace variable

# Compute the frequency response of the dead-time system
response = dead_time_system(s)

# Plot the frequency response
plt.figure(figsize=(10, 6))
plt.plot(frequency, np.abs(response), label='Magnitude Response')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.title('Frequency Response of Dead-Time System')
plt.legend()
plt.grid(True)
plt.show()
```
