```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Define the RC circuit dynamics
def rc_circuit(t, Vc, R, C, Vin):
    return (Vin(t) - Vc) / (R * C)

# Circuit parameters
R = 1.0  # Resistance (ohms)
C = 1.0  # Capacitance (farads)

# Input voltage function (step input at t=1)
def input_voltage(t):
    return 1.0 if t >= 1 else 0.0

# Initial condition (voltage across the capacitor at t=0)
Vc0 = 0.0

# Time span
t_span = [0, 5]

# Solve the differential equation using solve_ivp
sol = solve_ivp(lambda t, Vc: rc_circuit(t, Vc, R, C, input_voltage), t_span, [Vc0], t_eval=np.linspace(t_span[0], t_span[1], 1000))

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(sol.t, sol.y[0], label='Voltage across Capacitor')
plt.xlabel('Time')
plt.ylabel('Voltage (V)')
plt.title('RC Circuit Response to Step Input')
plt.legend()
plt.grid(True)
plt.show()
```
