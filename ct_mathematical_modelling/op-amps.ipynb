```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
R1 = 10e3  # Input resistor (10 kΩ)
Rf = 100e3  # Feedback resistor (100 kΩ)

# Frequency range
frequency = np.linspace(1, 100000, 1000)  # Frequency range from 1 Hz to 100 kHz

# Calculate gains for each frequency
gains = 1 + (Rf / R1) * np.ones_like(frequency)  # Generate an array of gains

# Plot the gain vs frequency
plt.figure(figsize=(10, 6))
plt.semilogx(frequency, 20 * np.log10(gains))  # Convert gain to decibels
plt.xlabel('Frequency (Hz)')
plt.ylabel('Gain (dB)')
plt.title('Frequency Response of Non-Inverting Operational Amplifier')
plt.grid(True)
plt.show()
```
