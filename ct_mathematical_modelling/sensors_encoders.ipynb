```python
import numpy as np
import matplotlib.pyplot as plt

# Constants (Steinhart-Hart coefficients)
A = 1.129241e-3
B = 2.341077e-4
C = 8.775468e-8

# Function to calculate temperature from resistance using Steinhart-Hart equation
def calculate_temperature(R):
    return 1 / (A + B * np.log(R) + C * (np.log(R))**3)

# Resistance values (simulating sensor readings)
resistances = np.linspace(1000, 10000, 1000)

# Calculate corresponding temperatures
temperatures = calculate_temperature(resistances)

# Plot the temperature vs. resistance curve
plt.figure(figsize=(10, 6))
plt.plot(resistances, temperatures, label='Temperature vs. Resistance')
plt.xlabel('Resistance (Ohms)')
plt.ylabel('Temperature (K)')
plt.title('Modeling Temperature Sensor Response')
plt.legend()
plt.grid(True)
plt.show()
```
