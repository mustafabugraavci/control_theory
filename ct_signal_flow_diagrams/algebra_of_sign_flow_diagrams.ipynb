```python
# Import required libraries
import control as ctrl
import matplotlib.pyplot as plt

# Define transfer functions for the blocks
G1 = ctrl.TransferFunction([1], [1, 2])  # Example transfer function 1
G2 = ctrl.TransferFunction([1], [1, 3])  # Example transfer function 2

# Series connection
G_series = ctrl.series(G1, G2)

# Parallel connection
G_parallel = ctrl.parallel(G1, G2)

# Feedback connection
H = ctrl.TransferFunction([1], [1, 4])  # Example feedback transfer function
T = ctrl.feedback(G1, H)

# Plot step response for each system
t, y1 = ctrl.step_response(G_series)
t, y2 = ctrl.step_response(G_parallel)
t, y3 = ctrl.step_response(T)

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(t, y1, label='Series Connection')
plt.plot(t, y2, label='Parallel Connection')
plt.plot(t, y3, label='Feedback Connection')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Step Response of Signal Flow Diagrams')
plt.legend()
plt.grid(True)
plt.show()
```
