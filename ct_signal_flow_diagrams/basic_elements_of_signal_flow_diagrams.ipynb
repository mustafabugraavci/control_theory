```python
# Define nodes and edges for the Signal Flow Diagram
nodes = {
    'Input': (0, 1),
    'G(s)': (1, 2),
    'Output': (2, 1)
}

edges = [('Input', 'G(s)', 'U(s)'), ('G(s)', 'Output', 'Y(s)')]

# Create a plot for the Signal Flow Diagram
plt.figure(figsize=(6, 4))

# Draw nodes
for node, (x, y) in nodes.items():
    plt.plot(x, y, 'bo', markersize=30)  # Draw nodes as blue circles
    plt.text(x, y, node, fontsize=12, ha='center', va='center')

# Draw edges
for start, end, label in edges:
    startX, startY = nodes[start]
    endX, endY = nodes[end]
    plt.arrow(startX, startY, endX - startX, endY - startY, color='b',
              shape='full', lw=0, length_includes_head=True, head_width=0.15)
    midX, midY = (startX + endX) / 2, (startY + endY) / 2
    plt.text(midX, midY, label, fontsize=12, ha='center', va='center', color='b')

# Set plot limits and display the diagram
plt.xlim(-1, 3)
plt.ylim(0, 3)
plt.axis('off')
plt.title('Signal Flow Diagram of a Control System')
plt.show()
```
