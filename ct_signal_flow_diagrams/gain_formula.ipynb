```python
# Import required libraries
import control as ctrl
import matplotlib.pyplot as plt

# Define transfer functions for the blocks
G1 = ctrl.TransferFunction([1], [1, 2])  # Example transfer function 1
G2 = ctrl.TransferFunction([1], [1, 3])  # Example transfer function 2
H = ctrl.TransferFunction([1], [1, 4])   # Example feedback transfer function

# Calculate forward path gain (Gf)
Gf = ctrl.series(G1, G2)

# Calculate loop gain (Gl)
Gl = ctrl.series(G2, H)

# Calculate overall transfer function (G) using the gain formula
G = Gf / (1 - Gl)

# Plot step response of the system
t, y = ctrl.step_response(G)

# Plotting the step response
plt.figure(figsize=(10, 6))
plt.plot(t, y)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Step Response of the Interconnected System')
plt.grid(True)
plt.show()
```
