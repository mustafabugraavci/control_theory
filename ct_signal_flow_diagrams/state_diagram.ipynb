```python
# Import required libraries
import control as ctrl
import matplotlib.pyplot as plt

# Define the state-space representation of the system
A = [[-1, -2], [3, -4]]
B = [[1], [0]]
C = [[0, 1]]
D = [[0]]

# Create a StateSpace object
system = ctrl.StateSpace(A, B, C, D)

# Convert to Transfer Function
tf = ctrl.ss2tf(system)

# Get State and Output Equations
states = system.A  # A matrix contains state equations coefficients
outputs = ctrl.ss2tf(system)  # Get transfer function representation for output equations

# Print State and Output Equations
print("State Equations:")
print(states)
print("Output Equations:")
print(outputs)
# Plot the step response of the system
t, y = ctrl.step_response(system)
plt.figure(figsize=(10, 6))
plt.plot(t, y)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Step Response of the System')
plt.grid(True)
plt.show()
```
