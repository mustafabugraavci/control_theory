```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# Define a basic transfer function
numerator = [1]  # Original numerator
denominator = [1, 3, 2]  # Original denominator

# Create the original system
original_system = signal.TransferFunction(numerator, denominator)

# Add a pole and create a new system
new_denominator = [1, 3, 2, 0]  # Adding a pole
pole_added_system = signal.TransferFunction(numerator, new_denominator)

# Add a zero and create another new system
new_numerator = [1, 1]  # Adding a zero
zero_added_system = signal.TransferFunction(new_numerator, denominator)

# Time vector for simulation
time = np.linspace(0, 10, 500)

# Plot the step response of the original system
time_orig, response_orig = signal.step(original_system, T=time)
plt.plot(time_orig, response_orig, label='Original System')

# Plot the step response with an added pole
time_pole, response_pole = signal.step(pole_added_system, T=time)
plt.plot(time_pole, response_pole, label='System with Added Pole')

# Plot the step response with an added zero
time_zero, response_zero = signal.step(zero_added_system, T=time)
plt.plot(time_zero, response_zero, label='System with Added Zero')

plt.title('Step Response Comparison')
plt.xlabel('Time (seconds)')
plt.ylabel('Response')
plt.legend()
plt.grid(True)
plt.show()
```
