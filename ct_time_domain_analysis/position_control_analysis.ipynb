```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# Define a third-order system with given coefficients
numerator = [2]  # Coefficients for the numerator of the transfer function
denominator = [1, 3, 3, 2]  # Coefficients for the denominator (third-order system)

# Create the system
system = signal.TransferFunction(numerator, denominator)

# Time vector for simulation
time = np.linspace(0, 10, 500)

# Compute and plot the step response
time_step, response_step = signal.step(system, T=time)
plt.figure()
plt.plot(time_step, response_step)
plt.title('Step Response of the Third-Order System')
plt.xlabel('Time (seconds)')
plt.ylabel('Response')
plt.grid(True)

# Compute and plot the ramp response
time_ramp, response_ramp = signal.step(system, T=time)
response_ramp = np.cumsum(response_ramp) * (time_ramp[1] - time_ramp[0])  # Integration for ramp response
plt.figure()
plt.plot(time_ramp, response_ramp)
plt.title('Ramp Response of the Third-Order System')
plt.xlabel('Time (seconds)')
plt.ylabel('Response')
plt.grid(True)
plt.show()
```
