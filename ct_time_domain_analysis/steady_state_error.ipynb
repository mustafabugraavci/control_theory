```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# System parameters
K = 1  # System gain
tau = 2  # Time constant

# Transfer function G(s) = K / (tau*s + 1)
numerator = [K]
denominator = [tau, 1]
system = signal.TransferFunction(numerator, denominator)

# Time parameters
t = np.linspace(0, 10, 500)

# Step response
t, response = signal.step(system, T=t)

# Calculate steady state error for a step input
desired_output = 1
steady_state_value = response[-1]
steady_state_error = np.abs(desired_output - steady_state_value)

# Plotting
plt.plot(t, response, label='System Response')
plt.axhline(y=steady_state_value, color='r', linestyle='--', label='Steady State Value')
plt.title('System Step Response')
plt.xlabel('Time')
plt.ylabel('Output')
plt.grid()
plt.legend()
plt.show()

# Output the steady state error
print(f"Steady State Error: {steady_state_error}")
```
