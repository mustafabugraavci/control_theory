```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# System parameters
m = 1.0  # Mass
k = 1.0  # Spring constant
c = 0.5  # Damping coefficient

# State equations
def mass_spring_damper(t, y):
    x1, x2 = y
    dx1dt = x2
    dx2dt = -(k/m) * x1 - (c/m) * x2
    return [dx1dt, dx2dt]

# Initial conditions
initial_state = [1.0, 0.0]  # Initial position and velocity

# Time span
t_span = (0, 10)

# Solve the ODE
sol = solve_ivp(mass_spring_damper, t_span, initial_state, t_eval=np.linspace(t_span[0], t_span[1], 1000))

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(sol.t, sol.y[0], label='Position (x1)')
plt.plot(sol.t, sol.y[1], label='Velocity (x2)')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Mass-Spring-Damper System Response')
plt.legend()
plt.grid(True)
plt.show()
```
